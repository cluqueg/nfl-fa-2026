# Combinar Reportes en Excel

Este notebook combina todos los CSV generados en un único archivo Excel con múltiples pestañas.

```python
import pandas as pd
from io import BytesIO

# Lista de archivos CSV a combinar (por posición)
csv_files = {
    'CB': 'cb.csv',
    'S': 's.csv',
    'LB': 'lb.csv',
    'DI': 'di.csv',
    'ED': 'ed.csv',
    'QB': 'qb.csv',
    'WR': 'wr.csv',
    'TE': 'te.csv',
    'RB': 'rb.csv',
    'OT': 'ot.csv',
    'IOL': 'iol.csv',
    'Posiciones': 'posiciones.csv'
}

output_path = './NFL_Free_Agency_2026.xlsx'

print(f"Combinando {len(csv_files)} reportes en Excel...")
```

```python
# Crear Excel writer
with pd.ExcelWriter(output_path, engine='openpyxl') as writer:
    for sheet_name, csv_file in csv_files.items():
        csv_path = f'./{csv_file}'
        
        try:
            # Leer CSV
            df = pd.read_csv(csv_path)
            
            # Escribir a Excel
            df.to_excel(writer, sheet_name=sheet_name, index=False)
            
            print(f"✓ {csv_file} -> Pestaña '{sheet_name}' ({len(df)} jugadores)")
        except FileNotFoundError:
            print(f"✗ No se encontró: {csv_file}")
        except Exception as e:
            print(f"✗ Error procesando {csv_file}: {str(e)}")

print(f"\n✓ Archivo Excel creado: {output_path}")
```

```python
# Verificar el archivo creado
import os
if os.path.exists(output_path):
    size = os.path.getsize(output_path)
    print(f"Archivo generado exitosamente:")
    print(f"  Ubicación: {output_path}")
    print(f"  Tamaño: {size:,} bytes")
else:
    print("Error: No se pudo crear el archivo")
```
