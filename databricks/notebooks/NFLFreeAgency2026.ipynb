# Agencia Libre de la NFL 2026

Este Notebook se usa para obtener informes relacionados con la Agencia Libre.

```python
%sql
-- Selecciona los mejores Defensas que son UFA
SELECT Pos, COUNT(Pos) AS AgentesLibres
  FROM freeagents_2026 as fa 
--  WHERE fa.Type IN ("UFA", "Option", "RFA", "ERFA", "SFA") -- UFA, RFA, ERFA, SFA, Void, Option
--  AND fa.Age < 30
  GROUP BY Pos
  ORDER BY AgentesLibres DESC
```

# Mejores CB

```python
%sql
-- Selecciona los mejores Defensas que son UFA
SELECT fa.Player, fa.Pos, fa.PrevTeam, fa.Age AS Edad,
-- Grades PFF
    pff.grades_defense AS DEF, pff.grades_run_defense AS RDEF, pff.grades_pass_rush_defense AS PRSH, pff.grades_coverage_defense AS COV,
    fa.`Snaps%`, fa.SalaryPY AS SalarioAnual

  FROM freeagents_2026 as fa 
  JOIN defense_summary AS pff 
  ON fa.Player = pff.player

  WHERE Type IN ("UFA", "SFA", "Void") -- UFA, RFA, ERFA, SFA, Void, Option
  AND fa.Pos in ("CB") -- "EDGE", "IDL", "LB", "S", "CB"
  --AND fa.Age < 29
  -- AND fa.`Snaps%` > 20
  --AND pff.grades_defense > 45
  SORT BY pff.grades_defense DESC
  -- LIMIT 10
```

# Mejores RB/FB

```python
%sql
SELECT fa.Player, fa.Pos, fa.PrevTeam , fa.Age,
-- Grades PFF
    pff.grades_offense AS OFF, pff.grades_run AS RUN, pff.grades_pass_block AS PBLK, pff.grades_run_block AS RBLK,
    fa.`Snaps%`, fa.SalaryPY 

  FROM freeagents_2026 as fa 
  JOIN rushing_summary AS pff 
  ON fa.Player = pff.player

  WHERE Type IN ("UFA", "Void") -- UFA, RFA, ERFA, SFA, Void, Option
  AND fa.Pos in ("RB", "FB") -- "QB", "WR", "LT", "RT". "TE", "RB", "FB" "LG", , "RG"
  -- AND fa.Age < 30
  -- AND fa.`Snaps%` > 10
  -- AND pff.grades_offense > 60
  SORT BY pff.grades_offense DESC
```

# Mejores Receptores

```python
%sql
SELECT fa.Player, fa.Pos, fa.PrevTeam , fa.Age,
-- Grades PFF
    pff.grades_offense AS OFF, pff.grades_pass_route AS RECV, pffblk.grades_pass_block AS PBLK, pffblk.grades_run_block AS RBLK,
    fa.`Snaps%`, fa.SalaryPY 

  FROM freeagents_2026 as fa 
  JOIN receiving_summary AS pff
  JOIN offense_blocking AS pffblk
  ON fa.Player = pff.player and fa.Player = pffblk.player
  WHERE Type IN ("UFA", "Void") -- UFA, RFA, ERFA, SFA, Void, Option
  AND fa.Pos in ("WR")
  -- AND fa.Age < 30
  -- AND fa.`Snaps%` > 10
  -- AND pff.grades_offense > 60
  SORT BY pff.grades_offense DESC
```

#Mejores TEs
